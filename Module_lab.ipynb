```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
days = [1,2,3,4,5]
amount = [1,5,20,450,700]
```

```python
plt.plot(days, amount,'bs')
plt.show()
```

```python
months = np.arange(1,13,1)
income = [100+3*month for month in months]
print('Income: {}'.format(income))
cost = [50+10*month for month in months]
print('Cost: {}'.format(cost))
```

```python
plt.plot(months,income,'go',months,cost,'r^')
plt.show()
```
